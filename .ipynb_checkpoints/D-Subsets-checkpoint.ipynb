# Subsetting AlienVault Dataset


**Strata considerations:**
In our case we use the ALIENVAULT collection - 2500 pulses. From these we keep only the 250 pulses which exist for 2020. For these we used the Quantile-based discretization function (pandas.qcut) to convert continuous variables into 4 quantiles.

In our sample we considered strata of different 'month' of the pulse, 'q_total_ind' (number of different indicators/IOCs which show the number of attacks from different IPs) and 'q_total_ref' (number of total external references)

**Notes for alternative approaches:**
   - We wanted to also create stratas based on 'q_total_mal_fam' (total mal familes) but the are quantiles were identical (1-1) with 'q_total_ref'. 
   - Another strata could be 'q_desc_word_count' (number of words in the description)
   - We did not find other diverging -meaningful- continuous variables to use them for different strata [we may consider this on the second iteration of labelling]

## Feature Extraction

```python
pulses_df['original_index'] = pulses_df.index.values

pulses_df['desc_word_count'] = pulses_df.description.str.split().map(lambda x: len(x) if isinstance(x, list) else np.nan)
pulses_df['q_desc_word_count'] = pd.qcut(pulses_df.desc_word_count, 4, duplicates='drop').apply(str)

pulses_df['total_ref'] = pulses_df.references.apply(lambda x: len(x))
pulses_df['q_total_ref'] = pd.qcut(pulses_df.total_ref, 4, duplicates='drop').apply(str)

pulses_df['total_attack_ids'] = pulses_df.attack_ids.apply(lambda x: len(x))
pulses_df['q_total_attack_ids'] = pd.qcut(pulses_df.total_attack_ids, 4, duplicates='drop').apply(str)

pulses_df['total_ind'] = pulses_df.indicators.apply(lambda x: len(x))
pulses_df['q_total_ind'] = pd.qcut(pulses_df.total_ind, 4).apply(str)

pulses_df['total_tags'] = pulses_df.tags.apply(lambda x: len(x))
pulses_df['q_total_tags'] = pd.qcut(pulses_df.total_tags, 4).apply(str)

pulses_df['total_mal_fam'] = pulses_df.malware_families.apply(lambda x: len(x))
pulses_df['q_total_mal_fam'] = pd.qcut(pulses_df.total_mal_fam, 4, duplicates='drop').apply(str)

pulses_df['year'] = pulses_df.created.dt.year
pulses_df['month'] = pulses_df.created.dt.month
```

# Stratified Subsets

## Definiting function

```python
#ref: https://github.com/flaboss/python_stratified_sampling/blob/master/stratifiedSample.py

# Functions
def stratified_sample(df, strata, size=None, seed=None, keep_index= True):
    '''
    It samples data from a pandas dataframe using strata. These functions use
    proportionate stratification:
    n1 = (N1/N) * n
    where:
        - n1 is the sample size of stratum 1
        - N1 is the population size of stratum 1
        - N is the total population size
        - n is the sampling size
    Parameters
    ----------
    :df: pandas dataframe from which data will be sampled.
    :strata: list containing columns that will be used in the stratified sampling.
    :size: sampling size. If not informed, a sampling size will be calculated
        using Cochran adjusted sampling formula:
        cochran_n = (Z**2 * p * q) /e**2
        where:
            - Z is the z-value. In this case we use 1.96 representing 95%
            - p is the estimated proportion of the population which has an
                attribute. In this case we use 0.5
            - q is 1-p
            - e is the margin of error
        This formula is adjusted as follows:
        adjusted_cochran = cochran_n / 1+((cochran_n -1)/N)
        where:
            - cochran_n = result of the previous formula
            - N is the population size
    :seed: sampling seed
    :keep_index: if True, it keeps a column with the original population index indicator
    
    Returns
    -------
    A sampled pandas dataframe based in a set of strata.
    Examples
    --------
    >> df.head()
    	id  sex age city 
    0	123 M   20  XYZ
    1	456 M   25  XYZ
    2	789 M   21  YZX
    3	987 F   40  ZXY
    4	654 M   45  ZXY
    ...
    # This returns a sample stratified by sex and city containing 30% of the size of
    # the original data
    >> stratified = stratified_sample(df=df, strata=['sex', 'city'], size=0.3)
    Requirements
    ------------
    - pandas
    - numpy
    '''
    population = len(df)
    size = __smpl_size(population, size)
    tmp = df[strata]
    tmp['size'] = 1
    tmp_grpd = tmp.groupby(strata).count().reset_index()
    tmp_grpd['samp_size'] = round(size/population * tmp_grpd['size']).fillna(0).astype(int)

    # controlling variable to create the dataframe or append to it
    first = True 
    for i in range(len(tmp_grpd)):
        # query generator for each iteration
        qry=''
        for s in range(len(strata)):
            stratum = strata[s]
            value = tmp_grpd.iloc[i][stratum]
            n = tmp_grpd.iloc[i]['samp_size']

            if type(value) == str:
                value = "'" + str(value) + "'"
            
            if s != len(strata)-1:
                qry = qry + stratum + ' == ' + str(value) +' & '
            else:
                qry = qry + stratum + ' == ' + str(value)
        
        # final dataframe
        if first:
            stratified_df = df.query(qry).sample(n=n, random_state=seed).reset_index(drop=(not keep_index))
            first = False
        else:
            tmp_df = df.query(qry).sample(n=n, random_state=seed).reset_index(drop=(not keep_index))
            stratified_df = stratified_df.append(tmp_df, ignore_index=True)
    
    return stratified_df


def stratified_sample_report(df, strata, size=None):
    '''
    Generates a dataframe reporting the counts in each stratum and the counts
    for the final sampled dataframe.
    Parameters
    ----------
    :df: pandas dataframe from which data will be sampled.
    :strata: list containing columns that will be used in the stratified sampling.
    :size: sampling size. If not informed, a sampling size will be calculated
        using Cochran adjusted sampling formula:
        cochran_n = (Z**2 * p * q) /e**2
        where:
            - Z is the z-value. In this case we use 1.96 representing 95%
            - p is the estimated proportion of the population which has an
                attribute. In this case we use 0.5
            - q is 1-p
            - e is the margin of error
        This formula is adjusted as follows:
        adjusted_cochran = cochran_n / 1+((cochran_n -1)/N)
        where:
            - cochran_n = result of the previous formula
            - N is the population size
    Returns
    -------
    A dataframe reporting the counts in each stratum and the counts
    for the final sampled dataframe.
    '''
    population = len(df)
    size = __smpl_size(population, size)
    tmp = df[strata]
    tmp['size'] = 1
    tmp_grpd = tmp.groupby(strata).count().reset_index()
    tmp_grpd['samp_size'] = round(size/population * tmp_grpd['size']).fillna(0).astype(int)
    return tmp_grpd


def __smpl_size(population, size):
    '''
    A function to compute the sample size. If not informed, a sampling 
    size will be calculated using Cochran adjusted sampling formula:
        cochran_n = (Z**2 * p * q) /e**2
        where:
            - Z is the z-value. In this case we use 1.96 representing 95%
            - p is the estimated proportion of the population which has an
                attribute. In this case we use 0.5
            - q is 1-p
            - e is the margin of error
        This formula is adjusted as follows:
        adjusted_cochran = cochran_n / 1+((cochran_n -1)/N)
        where:
            - cochran_n = result of the previous formula
            - N is the population size
    Parameters
    ----------
        :population: population size
        :size: sample size (default = None)
    Returns
    -------
    Calculated sample size to be used in the functions:
        - stratified_sample
        - stratified_sample_report
    '''
    if size is None:
        cochran_n = round(((1.96)**2 * 0.5 * 0.5)/ 0.02**2)
        n = round(cochran_n/(1+((cochran_n -1) /population)))
    elif size >= 0 and size < 1:
        n = round(population * size)
    elif size < 0:
        raise ValueError('Parameter "size" must be an integer or a proportion between 0 and 0.99.')
    elif size >= 1:
        n = size
    return n
```

```python
from langdetect import detect
#detect(temp.iloc[30].final)
def substring_after(s):
            
    if s != "":
        if detect(s) != 'en':   #remove non english documents
            s = ""    
        
    if 'Find out more on how we use cookies . Accept' in s:
        delim = 'Find out more on how we use cookies . Accept'
        s = s.partition(delim)[2]
        
    if 'Securelist Solutions' in s:
        delim = 'Transparency Center Policy Blog Corporate News Press Center Careers Incubator Sponsorships'
        s = s.partition(delim)[2]
    
    if 'About Cisco Login' in s:
        delim = 'About Cisco Login'
        s = s.partition(delim)[2] 
        
    if 'Critical Attack Discovery and Intelligence Team Symantec' in s:
        delim = 'Critical Attack Discovery and Intelligence Team Symantec'
        s = s.partition(delim)[2] 
        
    if 'My Account Cloud Console SUBSCRIBE Exploits |' in s:
        delim = 'My Account Cloud Console SUBSCRIBE Exploits |'
        s = s.partition(delim)[2] 
        
    if 'Vermont Virgin Islands Virginia Washington West Virginia Wisconsin Wyoming' in s:
        delim = 'Vermont Virgin Islands Virginia Washington West Virginia Wisconsin Wyoming'
        s = s.partition(delim)[2] 
        
    if 'Use SLUB Gets Rid of GitHub, Intensifies Slack Use Posted' in s:
        delim = 'Use SLUB Gets Rid of GitHub, Intensifies Slack Use Posted'
        s = s.partition(delim)[2] 
        
    if 'Contact Us Home > Blog >' in s:
        delim = 'Contact Us Home > Blog >'
        s = s.partition(delim)[2] 
        
    if 'About Us Careers Contact Us Sign in My Account Cloud Console' in s:
        delim = 'Contact Us Home > Blog >'
        s = s.partition(delim)[2] 
        
    if 'Main Menu You are here' in s:
        delim = 'Main Menu You are here'
        s = s.partition(delim)[2]
        
    if 'About Us Careers' in s:
        delim = 'About Us Careers'
        s = s.partition(delim)[2] 
        
    if 'Blog Contact Blog' in s:
        delim = 'About Us Careers'
        s = s.partition(delim)[2] 
        
    if 'Posted on' in s:
        delim = 'Posted on'
        s = s.partition(delim)[2]  
        
    if 'Page not found' in s:  #remove documents without content
        s = ""
        
    if 'Request unsuccessful' in s:
        s = ""        
        
    if 'CONTINUE TO SITE' in s:
        s = ""  
        
    if 'Risk Detected' in s:
        s = "" 
        
    if 'Skip to main content' in s:
        s = ""
        
        
    return s   

#substring_after(web_scraped.loc[web_scraped.ID==2320].final[254])
```

```python
web_scraped['final_updated'] = web_scraped.final.apply(lambda y: substring_after(y))
web_scraped['modified_web_scrape'] = web_scraped['final'] != web_scraped['final_updated']
```

```python
web_scraped['first_520'] = web_scraped.final_updated.apply(lambda x: ' '.join(x.split()[:520]))
web_scraped['final_520'] = web_scraped.final_updated.apply(lambda x: ' '.join(x.split()[-520:]))
```
