```python
nlp = spacy.load("en_core_web_sm")

def lf_preprocessing(DF):
    DF.description = DF.description.astype(str)
    DF['targeted_countries'] = DF['targeted_countries'].apply(lambda d: d if isinstance(d, list) else [])
    DF['targeted_countries_count'] = DF['targeted_countries'].apply(lambda x: len(x))
    DF['targeted_countries_ind'] = DF['targeted_countries'].apply(lambda x: True if len(x)>0 else False ).astype(int)

    DF['industries'] = DF['industries'].apply(lambda d: d if isinstance(d, list) else [])
    DF['industries_count'] = DF['industries'].apply(lambda x: len(x))
    DF['industries_ind'] = DF['industries'].apply(lambda x: True if len(x)>0 else False ).astype(int)

    DF['tags'] = DF['tags'].apply(lambda d: d if isinstance(d, list) else [])
    DF['tags_count'] = DF['tags'].apply(lambda x: len(x))
    DF['tags_ind'] = DF['tags'].apply(lambda x: True if len(x)>0 else False ).astype(int)

    DF['attack_ids'] = DF['attack_ids'].apply(lambda d: d if isinstance(d, list) else [])
    DF['attack_ids_count'] = DF['attack_ids'].apply(lambda x: len(x))
    DF['attack_ids_ind'] = DF['attack_ids'].apply(lambda x: True if len(x)>0 else False ).astype(int)

    DF['malware_families'] = DF['malware_families'].apply(lambda d: d if isinstance(d, list) else [])
    DF['malware_families_count'] = DF['malware_families'].apply(lambda x: len(x))
    DF['malware_families_ind'] = DF['malware_families'].apply(lambda x: True if len(x)>0 else False ).astype(int)
    
    DF['adversary_ind'] = DF.adversary.isnull().astype(int)
    
    ## Create lemmas for descriptions
    DF['description_lemma_spacy'] = DF['description'].apply(lambda x: [y.lemma_ for y in  nlp(x)])

    #DF.extract_source[DF.extract_source.notnull()] --> empty
    return DF
```

## A.3 Creation of LF for ['Targeted']

## DEFINITIONS

```python
ABSTAIN = -1
NEGATIVE = 0
POSITIVE = 1


from snorkel.labeling import LabelingFunction
import re

@labeling_function()
def NOUN_data(x):   
    return POSITIVE if re.search(r"data.*", x.description, flags=re.I) else NEGATIVE

@labeling_function()
def NOUN_spy(x):   
    return POSITIVE if re.search(r"spy.*", x.description, flags=re.I) else NEGATIVE

@labeling_function()
def NOUN_cyberespionage(x):   
    return POSITIVE if re.search(r"cyberespionage.*", x.description, flags=re.I) else NEGATIVE

@labeling_function()
def NOUN_espionage(x):   
    return POSITIVE if re.search(r"espionage.*", x.description, flags=re.I) else NEGATIVE

@labeling_function()
def PROPN_breach(x):   
    return POSITIVE if re.search(r"breach.*", x.description, flags=re.I) else NEGATIVE



lfs = [
       NOUN_data, NOUN_spy, NOUN_cyberespionage, NOUN_espionage,
    PROPN_breach
    
      ]
#lfs.append(length)
```

```python
lf_label = 'Espionage'
```

```python
lfs_experiment = lfs
len(lfs_experiment)
```

```python
names = pd.Series(lfs_experiment).apply(lambda x: x.name)
#print(', '.join(best_combinations))
```

```python
def powerset(s):
    x = len(s)
    masks = [1 << i for i in range(x)]
    for i in range(1 << x):
        yield [ss for mask, ss in zip(masks, s) if i & mask]
```

```python
names = list(powerset(pd.Series(lfs_experiment).apply(lambda x: x.name).to_list()))
powersets_list = list(powerset(lfs_experiment))
powersets = pd.Series(powersets_list).to_frame()
powersets.columns  =['LFs']
powersets['names'] = names
powersets['total_LFs'] = powersets['names'].apply(lambda x: len(x))
powersets = powersets.loc[powersets.total_LFs>2]
powersets = powersets.reset_index(drop=True)
powersets['AUC_ROC'] = 0
powersets.shape
powersets.head()
#powersets.loc[len(powersets.names)>2]
```

```python
for index, row in powersets.iterrows():
    temp_lfs = powersets.iloc[index].LFs
    applier = PandasLFApplier(lfs=temp_lfs)
    L_test = applier.apply(df=annot_gold)
    label_model = LabelModel(cardinality=2, verbose=True)
    label_model.fit(L_train=L_test, n_epochs=500, log_freq=100, seed=123)
    metrics = ['accuracy', 'coverage', 'precision', 'recall', 'f1', 'roc_auc']
    generative_model_scores = label_model.score(L=L_test, Y=annot_gold[lf_label], metrics=metrics, tie_break_policy="random")
    powersets.loc[index,['AUC_ROC']] = round(generative_model_scores['roc_auc'],3)
        
powersets = powersets.sort_values(by='AUC_ROC', ascending=False).reset_index(drop=True)
powersets.head()
```

```python
powersets = powersets.sort_values(by='AUC_ROC', ascending=False).reset_index(drop=True)
powersets.head()


powersets_viz = powersets.loc[:,['total_LFs', 'AUC_ROC']].reset_index(drop=True)
ax = powersets_viz.plot.scatter(x='total_LFs', y='AUC_ROC', yticks =[0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]  ,xticks=[3,4,5], figsize=(8,6))
ax.set_title('AUC over different number of Labelling Functions \n(Label:Espionage - Short Descriptions)')
ax.set_xlabel('Labelling Functions')
ax.set_ylabel('Area Under the Curve')
plt.show();
```

```python
from snorkel.augmentation import transformation_function

from snorkel.preprocess.nlp import SpacyPreprocessor

spacy = SpacyPreprocessor(text_field="text", doc_field="doc", memoize=True)

# Pregenerate some random person names to replace existing ones with
# for the transformation strategies below


# Replace a random named entity with a different entity of the same type.
@transformation_function(pre=[spacy])
def change_country(x):
    person_names = [ent.text for ent in x.doc.ents if ent.label_ == "GPE"]
    # If there is at least one person name, replace a random one. Else return None.
    if person_names:
        name_to_replace = np.random.choice(person_names)
        replacement_countries = np.random.choice(replacement_countries_ISO)
        x.text = x.text.replace(name_to_replace, replacement_countries)
        return x


# Swap two adjectives at random.
@transformation_function(pre=[spacy])
def swap_adjectives(x):
    adjective_idxs = [i for i, token in enumerate(x.doc) if token.pos_ == "ADJ"]
    # Check that there are at least two adjectives to swap.
    if len(adjective_idxs) >= 2:
        idx1, idx2 = sorted(np.random.choice(adjective_idxs, 2, replace=False))
        # Swap tokens in positions idx1 and idx2.
        x.text = " ".join(
            [
                x.doc[:idx1].text,
                x.doc[idx2].text,
                x.doc[1 + idx1 : idx2].text,
                x.doc[idx1].text,
                x.doc[1 + idx2 :].text,
            ]
        )
        return x
```

```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download("wordnet")

def get_synonym(word, pos=None):
    """Get synonym for word given its part-of-speech (pos)."""
    synsets = wn.synsets(word, pos=pos)
    # Return None if wordnet has no synsets (synonym sets) for this word and pos.
    if synsets:
        words = [lemma.name() for lemma in synsets[0].lemmas()]
        if words[0].lower() != word.lower():  # Skip if synonym is same as word.
            # Multi word synonyms in wordnet use '_' as a separator e.g. reckon_with. Replace it with space.
            return words[0].replace("_", " ")


def replace_token(spacy_doc, idx, replacement):
    """Replace token in position idx with replacement."""
    return " ".join([spacy_doc[:idx].text, replacement, spacy_doc[1 + idx :].text])

@transformation_function(pre=[spacy])
def replace_verb_with_synonym(x):
    # Get indices of verb tokens in sentence.
    verb_idxs = [i for i, token in enumerate(x.doc) if token.pos_ == "VERB"]
    if verb_idxs:
        # Pick random verb idx to replace.
        idx = np.random.choice(verb_idxs)
        synonym = get_synonym(x.doc[idx].text, pos="v")
        # If there's a valid verb synonym, replace it. Otherwise, return None.
        if synonym:
            x.text = replace_token(x.doc, idx, synonym)
            return x


@transformation_function(pre=[spacy])
def replace_noun_with_synonym(x):
    # Get indices of noun tokens in sentence.
    noun_idxs = [i for i, token in enumerate(x.doc) if token.pos_ == "NOUN"]
    if noun_idxs:
        # Pick random noun idx to replace.
        idx = np.random.choice(noun_idxs)
        synonym = get_synonym(x.doc[idx].text, pos="n")
        # If there's a valid noun synonym, replace it. Otherwise, return None.
        if synonym:
            x.text = replace_token(x.doc, idx, synonym)
            return x
        
        

@transformation_function(pre=[spacy])
def replace_adjective_with_synonym(x):
    # Get indices of adjective tokens in sentence.
    adjective_idxs = [i for i, token in enumerate(x.doc) if token.pos_ == "ADJ"]
    if adjective_idxs:
        # Pick random adjective idx to replace.
        idx = np.random.choice(adjective_idxs)
        synonym = get_synonym(x.doc[idx].text, pos="a")
        # If there's a valid adjective synonym, replace it. Otherwise, return None.
        if synonym:
            x.text = replace_token(x.doc, idx, synonym)
            return x
```

```python
#Defining functions to generate sliding windows

#position only
# def split_sets(full_text, n_splits):
#     total_words = len(full_text.split())
#     splits = [round((total_words/n_splits)*i) for i in range(1,n_splits)]
#     # split.append([split[16][1], xx])
#     splits = splits[::-1]
#     splits.append(0)
#     splits = splits[::-1]
#     #split
#     comb_split = [[splits[i-1], splits[i+1]] for i in range(1,n_splits-1)]
#     comb_split.append([comb_split[n_splits-4][1], total_words])
#     return comb_split
#[x ** 2 for x in range(7) if x % 2 == 0]

#actual windows
def split_sets_to_list(full_text, n_splits):
    total_words = len(full_text.split())
    splits = [round((total_words/n_splits)*i) for i in range(1,n_splits)]
    # split.append([split[16][1], xx])
    splits = splits[::-1]
    splits.append(0)
    splits = splits[::-1]
    #split
    comb_split = [[splits[i-1], splits[i+1]] for i in range(1,n_splits-1)]
    comb_split.append([comb_split[n_splits-4][1], total_words])
    text_splits = []
    for i in range(0,n_splits-1):
        text_iter = ' '.join(full_text.split()[ comb_split[i][0]  : comb_split[i][1]  ])
        text_splits.append(text_iter)
    return text_splits

#apply in the df (extends the previous df)
def sliding_windows(df, n_splits, threshold):
    text_series = df.final_updated
    ids = list(df.ID)
    
    splits_df = df.ID.to_frame()
    splits_df = splits_df.reindex(columns=list(range(0,n_splits-1)))
    splits_df = splits_df.add_prefix('Text Window ')
    #splits.loc[1180] = 50
    for i, ID in enumerate(ids):
            if len(text_series[ID].split()) <= threshold:
                splits_df.loc[ID] = text_series[ID]
            else:
                splits_df.loc[ID] = split_sets_to_list(text_series[ID], n_splits) 
    df = df.drop('ID', axis=1)
    final_df = df.merge(splits_df.reset_index(), on='ID', how='left')
    return final_df
```
